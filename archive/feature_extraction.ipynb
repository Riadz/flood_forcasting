```python
import math
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
from datetime import datetime
from Models import AutoEncoder
from utils import toTorchTensers, NormalizeData

base_url = r'.'

data_url = rf'{base_url}/data/Data.xlsx'
data_egypt = pd.read_excel(data_url, sheet_name='Egypt')
data_vietnam = pd.read_excel(data_url, sheet_name='Vietnam')

device = 'cpu'
# device = 'cuda' if torch.cuda.is_available() else 'cpu'
device = torch.device(device)
```

```python
# data
DATA = data_egypt
DATA = NormalizeData(DATA)
DATA = DATA.to_numpy()

DATA_X = DATA[:, 1:]
DATA_Y = DATA[:, 0]

[DATA_X, DATA_Y] = toTorchTensers(DATA_X, DATA_Y, device=device)
```

```python
loaded_model = AutoEncoder().to(device)
loaded_model.load_state_dict(
    torch.load(f'{base_url}/models/X2_auto_model_06-04-2022_23-27-39.pt')
)
loaded_model.eval()
print('model loaded')
```

```python
new_feats = []
for input in DATA_X:
  encoded = loaded_model.encoder(input)
  encoded = encoded.detach().cpu().numpy()
  new_feats.append(encoded)

print(len(new_feats))
```

```python
new_feats = np.array(new_feats)
DATA = pd.DataFrame({
  'Flood': DATA_Y,
  'Feature_1': new_feats[:, 0],
  'Feature_2': new_feats[:, 1],
  'Feature_3': new_feats[:, 2],
})

print(DATA)

current_datetime = datetime.now().strftime("%d-%m_%H-%M-%S")
path = rf'{base_url}/data/New_Feats_{current_datetime}.xlsx'
DATA.to_excel(path)
print(f'✅ data saved to "New_Feats_{current_datetime}.xlsx"')
```
